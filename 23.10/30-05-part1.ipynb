```python
import pandas as pd
```

```python
regs = pd.read_csv("https://raw.githubusercontent.com/allatambov/pydat19-ap/master/30-05/elect_clean.csv",
                  encoding = "UTF-8")
```

```python
for name, data in regs.groupby('region'):
    data.to_csv(name + ".csv") 
```

```python
def my_range(x):
    return max(x) - min(x) 
```

```python
regs.groupby('region').agg(my_range) 
```

```python
regs.groupby('region').agg(lambda x: max(x) - min(x)) 
```

```python
tot = regs.groupby('region').agg('sum')
```

```python
tot['turnout'] = tot['invalid'] + tot['valid'] 
```

```python
tot['turnout_perc'] = tot['turnout'] / tot['total'] * 100
```

```python
def to_perc(x):
    return x / tot['turnout'] * 100
```

```python
tot_perc = tot[['Zh', 'Zu', 'Mi', 
                'Pr', 'Pu']].apply(to_perc)
```

```python
new = [col + "_perc" for col in list(tot_perc.columns)]
tot_perc.columns = new
```

```python
full = pd.concat([tot, tot_perc], axis=1) 
```

```python
import matplotlib
```

```python
%matplotlib inline
```

```python
final = full.iloc[:, 10:16] 
```

```python
ax = final.plot.scatter('turnout_perc', 'Pu_perc', 
                        color = 'magenta') 
ax.set_title('Диаграмма рассеивания')
ax.set_xlabel('Явка (в %)')
ax.set_ylabel('Число голосов за Путина (в %) ') 
```
